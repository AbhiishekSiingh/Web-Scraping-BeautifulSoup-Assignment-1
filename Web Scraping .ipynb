# c) Top 10 ODI bowlers along with the records of their team and rating.

```python
page5=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
```

```python
soup=BeautifulSoup(page5.content)
soup
```

```python

bowler = soup.find('table', {'class': 'table rankings-table'})

players = []
teams = []
ratings = []

for row in bowler.find_all('tr')[1:11]:  
    columns = row.find_all('td')
    player = columns[1].text.strip()
    team = columns[2].text.strip()
    rating = columns[3].text.strip()
    players.append(player)
    teams.append(team)
    ratings.append(rating)

df = pd.DataFrame({'Player': players, 'Team': teams, 'Rating': ratings})
df
```

# 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name  
ii) Cuisine  
iii) Location  
iv) Ratings  
v) Image URL  

```python
page12=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page12
```

```python
soup=BeautifulSoup(page12.content)
soup
```

```python
restaurant_name=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    restaurant_name.append(i.text)
restaurant_name
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
cuisine
```

```python
import re

data = ['₹ 2,000 for 2 (approx) | Chinese, North Indian',
        '₹ 3,000 for 2 (approx) | Italian, Continental',
        '₹ 2,400 for 2 (approx) | North Indian, Italian',
        '₹ 1,700 for 2 (approx) | North Indian, Chinese',
        '₹ 1,800 for 2 (approx) | North Indian',
        '₹ 1,900 for 2 (approx) | North Indian',
        '₹ 2,200 for 2 (approx) | North Indian, Mughlai',
        '₹ 1,500 for 2 (approx) | North Indian, Continental, Chinese, South Indian']


cuisine = []
for item in data:
    
    pattern = r'\| (.*)'
    match = re.search(pattern, item)
    
   
    if match:
        cuisine_info = match.group(1)
        cuisine.append(cuisine_info)
    else:
        cuisine.append(item.text.strip())  

print(cuisine)
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location
```

```python
rating=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating.append(i.text)
rating
```

```python
image_url=[]
for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
image_url
```

```python
print(len(restaurant_name),len(cuisine),len(location),len(rating),len(image_url))
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant_Name':restaurant_name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image_URL':image_url})
df
```
